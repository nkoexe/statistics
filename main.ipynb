### Variables in the dataset

Information about the respondant and their neighbourhood:

* **ID** a unique number for each participant
* **Sex**
* **Age** in years

* **Ethnic Group** the categories given are the ones recorded in the original survey

### First dataset analysis
We can start looking at the ranges and type of data we are working with by using `crime.info()`.

```python
crime.info()
```

This shows us that all of the 9300 entries of the dataset have all non-null values, meaning we will have to only worry about filler values.
`crime.head()` can give us an overview of the how the data is encoded:

```python
crime.head()
```

We notice for example that `confpolice` and `antisoc` have seemingly arbitrary ranges, not normalized to the -1:1 range. We can already deduce that the `99` values of the fourth row are likely filler values, but to check for obvious filler outliers we can make use of `crime.describe()`.

```python
crime.describe()
```

We notice that the 75th centile of the columns `DeprivationIndex`, `confpolice` and `antisoc` are very far from the maximum value, which is `99` in all cases. We can check all these numerical columns using plotting functions to see how the data aggregates.

```python
numeric_cols = ["Age", "DeprivationIndex", "confpolice", "antisoc"]

for col in numeric_cols:
    sns.histplot(x=crime[col])
    plt.title(col)
    plt.show()
```

We can now be sure about `99` being a filler value in `DeprivationIndex`, `confpolice` and `antisoc`, but what about `Age`? People older than 95 are present in the dataset, as shown by

```python
sorted(crime.Age.unique())[-5:]
```

However, the number of people surveyed who are supposedly of 99 years of age still stands out.

```python
for i in range(90, 100):
    print(f"{i}-year olds: {crime.Age[crime.Age == i].count()}")
```

We decided therefore to treat the `99` value in the `Age` column as a filler value, like other previously mentioned numerical columns.

### Replacing numerical filler values
Now that we know which values need to be subsituted, we make use of the `.replace` function to insert `NaN` values instead.

```python
crime.Age.replace({99: np.nan}, inplace=True)
crime.DeprivationIndex.replace({99: np.nan}, inplace=True)
crime.confpolice.replace({99: np.nan}, inplace=True)
crime.antisoc.replace({99: np.nan}, inplace=True)
```

### Indetifying string filler values

**HINT** as `Victim<` is coded as 1 (if they have been a victim of crime in the past 12 months) and 0 (otherwise), you can obtain the proportion of people who have been a victim by taking the mean value of the column `Victim`.

You can also use `sns.barplot()` with the `x` and `hue` arguments to plot the proportion who are victims of crime within each category (each age group, etc).


```python
crime.groupby('Sex')['Victim'].mean()

# Plot
sns.barplot(data=crime, x='Sex', y='Victim')
plt.title('Proportion of Crime Victims by Sex')
plt.ylabel('Proportion Victim (mean of Victim)')
plt.show()
```

In proportion there was a very small difference between the likeliness of being a victim in the two sexes, with men being marginally (negligibly) more likely to experience crime.

```python
sns.barplot(data=crime, x='AgeGroup', y='Victim')
plt.title('Proportion of Victims by Age Group')
plt.ylabel('Proportion Victim')
plt.show()

crime.groupby('AgeGroup')['Victim'].mean()
```

The graph shows a correlation between age and likeliness of being a victim: younger people (<25) experienced crime more often (1 every 4) compared to older people (>75, 1 every 20).

```python
sns.barplot(data=crime, x='SES', y='Victim')
plt.title('Victimization Rate by Socio-Economic Status')
plt.ylabel('Proportion Victim')
plt.show()

crime.groupby('SES')['Victim'].mean()
```

Perhaps linked to the age demographic, the 7th group (full-time students) has the greatest relative amount of past victims of crime. Group 6 (unemployed) and 8 (unknown) also stand out in victim percentage.

```python
plt.figure(figsize=(10,5))
sns.barplot(data=crime, x='EthnicGroup', y='Victim')
plt.title('Proportion of Victims by Ethnic Group')
plt.xticks(rotation=45)
plt.ylabel('Proportion Victim')
plt.show()

crime.groupby('EthnicGroup')['Victim'].mean().sort_values(ascending=False)
```

Chinese, white and 'other' ethnic groups are sunject to crime the least amount compared to other groups, with a bigger percentage present in the 'mixed' group.

```python
sns.barplot(data=crime, x='DeprivationIndex', y='Victim')
plt.title('Victimization by Neighbourhood Deprivation Quintile')
plt.ylabel('Proportion Victim')
plt.xlabel('Deprivation Quintile (1 = Most Deprived, 5 = Least Deprived)')
plt.show()

crime.groupby('DeprivationIndex')['Victim'].mean()
```

The graph here shows a clear correlation between the average neighbourhood wealth data and criminality. Wealthier neighbourhoods experience less crime, whereas in poorer areas the chance of being in a difficult position is higher, resulting in a higher criminality rate.
